# <br>__1. Data: Pima Indians Diabetes Database__

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  # data split
```

### Step 1. 데이터 불러오기

```python
# 데이터 불러오기
data = pd.read_csv('./data/pima-indians-diabetes.data.csv', header=None)
data.head()
```

```python
# 변수명 변경
data.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', \
                'DiabetesPedigreeFunction', 'Age', 'Class']
data.head()
```

```python
X = data.iloc[:, :-1]  # 0-7 column은 독립변수
X.head()
```

```python
y = data.iloc[:, -1]  # 마지막 column은 종속변수
y.head()
```

### Step 2. 데이터 Split

```python
# 데이터 분할 train:test = 7:3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('Number of train set:', len(X_train))
print('Number of test set:', len(X_test))
```

---

# __2. Ensemble Model: Gradient Boosting Model Update Versions__

# 2-1. XGBoost

```python
import xgboost as xgb
from xgboost import plot_importance
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

- XGBClassifier 설명
    - objective: objective function ('reg:linear', 'binary:logistic', 'multi:softmax', 'multi:softprob'). **디폴트 'binary:logistic'**
    - learning_rate: 디폴트 0.3
    - max_depth: 각 의사결정나무의 최대 깊이. 디폴트 6
    - n_estimators: 모형 갯수. 디폴트 100
    - **reg_lambda: 디폴트 1**

### Step 1. 모델 선언 및 학습

```python
xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)
xgb_model.fit(X_train, y_train)
```

### Step 2. 예측 및 모델 평가

```python
xgb_y_pred = xgb_model.predict(X_test)
xgb_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7359)
xgb_acc = accuracy_score(y_true=y_test, y_pred=xgb_y_pred)
xgb_acc
```

==> XGBoost 모델이 base 모델보다 향상된 성능을 도출함
<br><br>

### Step 3. Hyperparameter 탐색

```python
# Hyperparameters 후보 설정
xgb_param_grid = {"learning_rate": [0.5, 0.3, 0.1],
                  "max_depth": [4, 6, 8],
                  "n_estimators": [50, 100, 150]
                 }
```

```python
# Hyperparameter 탐색 진행
xgb_model = xgb.XGBClassifier(objective='binary:logistic', seed=42)

xgb_grid_search = GridSearchCV(xgb_model, param_grid=xgb_param_grid, cv=5,scoring="accuracy")
xgb_grid_search.fit(X_train, y_train)
```

```python
# 최적 hyperparameter 확인 및 최적 모델 구축
xgb_grid_search.best_params_
```

```python
xgb_opt_model = xgb_grid_search.best_estimator_
xgb_opt_model
```

### Step 4. 예측 및 모델 평가

```python
# Step4. 예측
xgb_opt_y_pred = xgb_opt_model.predict(X_test)
xgb_opt_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7489)
xgb_opt_acc = accuracy_score(y_true=y_test, y_pred=xgb_opt_y_pred)
xgb_opt_acc
```

### Step 5. 변수 중요도 확인

```python
xgb_opt_model.feature_importances_
```

```python
xgb_opt_var_imp = pd.Series(xgb_opt_model.feature_importances_, index=data.columns[:-1])
xgb_opt_var_imp.sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plot_importance(xgb_opt_model, ax=ax)
```

# __<br>2-2. LightGBM__

```python
import lightgbm as lgb
from lightgbm import plot_importance
import matplotlib.pyplot as plt
```

- LGBMClassifier 설명
    - boosting_type: ‘gbdt’-traditional Gradient Boosting Decision Tree. ‘dart’-Dropouts meet Multiple Additive Regression Trees. ‘goss’-Gradient-based One-Side Sampling. ‘rf’-Random Forest. 디폴트 ‘gbdt’.
    - learning_rate: 디폴트 0.1
    - max_depth: 각 의사결정나무의 최대 깊이. 디폴트 -1 (no limit)
    - n_estimators: 모형 개수. 디폴트 100
    - objective: objective function. ‘regression’ for LGBMRegressor, ‘binary’ or ‘multiclass’ for LGBMClassifier, ‘lambdarank’ for LGBMRanker. 디폴트 None

### Step 1. 모델 선언 및 학습

```python
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(X_train, y_train)
```

### Step 2. 예측 및 모델 평가

```python
lgb_y_pred = lgb_model.predict(X_test)
lgb_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7273)
lgb_acc = accuracy_score(y_true=y_test, y_pred=lgb_y_pred)
lgb_acc
```

==> LightGBM 모델이 base 모델보다 향상된 성능을 도출함
<br><br>

### Step 3. Hyperparameter 탐색

```python
# Hyperparameters 후보 설정
lgb_param_grid = {'max_depth': [5, 10, 15],
                  'learning_rate': [1, 0.1, 0.01],
                  'n_estimators': [50, 100, 150]
                 }
```

```python
# Hyperparameter 탐색 진행
lgb_model = lgb.LGBMClassifier(random_state=42)

lgb_grid_search = GridSearchCV(lgb_model, param_grid=lgb_param_grid, cv=5, scoring='accuracy')
lgb_grid_search.fit(X_train, y_train)
```

```python
# 최적 hyperparameter 확인 및 최적 모델 구축
lgb_grid_search.best_params_
```

```python
lgb_opt_model = lgb_grid_search.best_estimator_
lgb_opt_model
```

### Step 4. 예측 및 모델 평가

```python
lgb_opt_y_pred = lgb_opt_model.predict(X_test)
lgb_opt_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7489)
lgb_opt_acc = accuracy_score(y_true=y_test, y_pred=lgb_opt_y_pred)
lgb_opt_acc
```

### Step 5. 변수 중요도 확인

```python
lgb_opt_model.feature_importances_
```

```python
lgb_opt_var_imp = pd.Series(lgb_opt_model.feature_importances_, index=data.columns[:-1])
lgb_opt_var_imp.sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plot_importance(lgb_opt_model, ax=ax)
```

# __<br>2-3. CatBoost__

```python
import catboost as cb
```

- CatBoostClassifier 설명
    - iterations: 의사결정나무 개수. 디폴트 1000
    - learning_rate: 디폴트 0.3
    - depth: 각 의사결정나무의 최대 깊이. 디폴트 6

### Step 1. 모델 선언 및 학습

```python
cb_model = cb.CatBoostClassifier(iterations=100, random_state=42)
cb_model.fit(X_train, y_train)
```

### Step 2. 예측 및 모델 평가

```python
cb_y_pred = cb_model.predict(X_test)
cb_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7706)
cb_acc = accuracy_score(y_true= y_test, y_pred= cb_y_pred)
cb_acc
```

==> CatBoost 모델이 base 모델보다 향상된 성능을 도출함
<br><br>

### Step 3. Hyperparameter 탐색

```python
# Hyperparameters 후보 설정
cb_param_grid = {'depth': [3, 6, 9],
                 'iterations': [50, 100, 150]
                }
```

```python
# Hyperparameter 탐색 진행
cb_model = cb.CatBoostClassifier(random_state=42)

cb_grid_search = GridSearchCV(cb_model, param_grid=cb_param_grid, cv=5, scoring='accuracy')
cb_grid_search.fit(X_train, y_train)
```

```python
# 최적 hyperparameter 확인 및 최적 모델 구축
cb_grid_search.best_params_
```

```python
cb_opt_model = cb_grid_search.best_estimator_
cb_opt_model
```

### Step 4. 예측 및 모델 평가

```python
cb_opt_y_pred = cb_opt_model.predict(X_test)
cb_opt_y_pred
```

```python
# 테스트 데이터에 대한 성능 평가 (=0.7619)
cb_opt_acc = accuracy_score(y_true=y_test, y_pred=cb_opt_y_pred)
cb_opt_acc
```

### Step 5. 변수 중요도 확인

```python
cb_opt_model.feature_importances_
```

```python
cb_opt_var_imp = pd.Series(cb_opt_model.feature_importances_, index=data.columns[:-1])
cb_opt_var_imp.sort_values(ascending=False)
```
